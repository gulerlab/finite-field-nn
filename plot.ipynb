```python
import matplotlib.pyplot as plt
```

```python
acc_arr = []
loss_arr = []
with open('./nets/real_net/062223_pinet_lenet_mnist.txt', 'r') as fp:
    for line in fp:
        accuracy, loss = line[:-1].split(', ')[-2], line[:-1].split(', ')[-1]
        acc_arr.append(float(accuracy.split(' ')[-1]))
        loss_arr.append(float(loss.split(' ')[-1]))

plt.figure()
plt.plot(list(range(len(acc_arr))), acc_arr)
plt.title('Real Domain - PiNet - MNIST - Accuracy')
plt.xlabel('iteration')
plt.ylabel('accuracy')
plt.show()

plt.figure()
plt.plot(list(range(len(loss_arr))), loss_arr)
plt.title('Real Domain - PiNet - MNIST - Loss')
plt.xlabel('iteration')
plt.ylabel('loss')
plt.show()
```

```python
acc_arr = []
loss_arr = []
with open('./nets/finite_field_net/062223_pinet_lenet_mnist.txt') as fp:
    for line in fp:
        if line[0] == 'e' and line[-2] != '0':
            loss_arr.append(float((line[:-1].split(', ')[-1]).split(' ')[-1]))

        if line[0] == 'e' and line[-2] == '0':
            acc_arr.append(float((line[:-1].split(', ')[-2]).split(' ')[-1]))

plt.figure()
plt.plot(list(range(len(acc_arr))), acc_arr)
plt.title('Finite Field Domain - PiNet - MNIST - Accuracy')
plt.xlabel('iteration')
plt.ylabel('accuracy')
plt.show()

plt.figure()
plt.plot(list(range(len(loss_arr))), loss_arr)
plt.title('Finite Field Domain - PiNet - MNIST - Loss')
plt.xlabel('iteration')
plt.ylabel('loss')
plt.show()
```
