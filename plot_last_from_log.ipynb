```python
import matplotlib.pylab as plt

# MNIST
## mnist clover finite field
mnist_ff_accuracy = []
with open('logs/2024329-mnist-v2-ff-clover.log', 'r') as fp:
    for line in fp:
        if ', acc:' in line:
            mnist_ff_accuracy.append(float(line.split(' ')[-1].split('#')[0]))


## mnist clover baseline
mnist_baseline_accuracy = []
with open('logs/2024329-mnist-clover-baseline-bg.log', 'r') as fp:
    for line in fp:
        if 'accuracy:' in line:
            mnist_baseline_accuracy.append(float(line.split(' ')[-1]))

## mnist clover baseline
mnist_baseline_accuracy_fg = []
with open('logs/2024329-mnist-clover-baseline-fg.log', 'r') as fp:
    for line in fp:
        if 'accuracy:' in line:
            mnist_baseline_accuracy_fg.append(float(line.split(' ')[-1]))

fig, ax = plt.subplots(figsize=(8,6), dpi=600)
for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(2)
    ax.spines[axis].set_color("black")
    ax.spines[axis].set_zorder(0)
plt.plot(mnist_ff_accuracy,color='b', linestyle='-', marker='^',markevery=2, markersize=10)
plt.plot(mnist_baseline_accuracy,color='r',linestyle='-', markevery=2, marker='X', markersize=10)
plt.plot(mnist_baseline_accuracy_fg,color='g',linestyle='-', markevery=2, marker='>', markersize=10)
plt.xlabel('# Iterations\n',fontweight='bold', fontsize=24)
plt.ylabel('Accuracy',fontweight='bold', fontsize=24)
locs, labels = plt.xticks()
plt.xticks(locs[1:-1], ['0', '100', '200', '300'])
plt.grid()
plt.rc('xtick', labelsize=23) 
plt.rc('ytick', labelsize=23) 
lgnd = plt.legend(['CLOVER', 'Conventional Neural Network (Batch GD)', 'Conventional Neural Network (Full GD)'], fontsize=16)
plt.savefig("mnist_clover_baseline_acc_with_replacement.png", bbox_inches = 'tight')
plt.savefig("mnist_clover_baseline_acc_with_replacement.eps", format='eps', bbox_inches = 'tight')
plt.show()

## cifar10_vgg clover finite field
cifar10_vgg_ff_accuracy = []
with open('logs/2024329-cifar10-vgg-v2-ff-clover.log', 'r') as fp:
    for line in fp:
        if ', acc:' in line:
            cifar10_vgg_ff_accuracy.append(float(line.split(' ')[-1].split('#')[0]))


## cifar10_vgg clover baseline
cifar10_vgg_baseline_accuracy = []
with open('logs/2024329-cifar10-vgg-clover-baseline-bg.log', 'r') as fp:
    for line in fp:
        if 'accuracy:' in line:
            cifar10_vgg_baseline_accuracy.append(float(line.split(' ')[-1]))

## cifar10_vgg clover baseline
cifar10_vgg_baseline_accuracy_fg = []
with open('logs/2024329-cifar10-vgg-clover-baseline-fg.log', 'r') as fp:
    for line in fp:
        if 'accuracy:' in line:
            cifar10_vgg_baseline_accuracy_fg.append(float(line.split(' ')[-1]))

fig, ax = plt.subplots(figsize=(8,6), dpi=600)
for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(2)
    ax.spines[axis].set_color("black")
    ax.spines[axis].set_zorder(0)
plt.plot(cifar10_vgg_ff_accuracy,color='b', linestyle='-', marker='^',markevery=2, markersize=10)
plt.plot(cifar10_vgg_baseline_accuracy,color='r',linestyle='-', markevery=2, marker='X', markersize=10)
plt.plot(cifar10_vgg_baseline_accuracy_fg,color='g',linestyle='-', markevery=2, marker='>', markersize=10)
plt.xlabel('# Iterations\n',fontweight='bold', fontsize=24)
plt.ylabel('Accuracy',fontweight='bold', fontsize=24)
locs, labels = plt.xticks()
plt.xticks(locs[1:-1], ['0', '100', '200', '300'])
plt.grid()
plt.rc('xtick', labelsize=23) 
plt.rc('ytick', labelsize=23) 
lgnd = plt.legend(['CLOVER', 'Conventional Neural Network (Batch GD)', 'Conventional Neural Network (Full GD)'], fontsize=16)
plt.savefig("cifar10_vgg_clover_baseline_acc_with_replacement.png", bbox_inches = 'tight')
plt.savefig("cifar10_vgg_clover_baseline_acc_with_replacement.eps", format='eps', bbox_inches = 'tight')
plt.show()
```
