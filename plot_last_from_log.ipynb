```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
# MNIST
## mnist clover finite field
mnist_ff_loss = []
mnist_ff_accuracy = []
with open('logs/202431-144112-mnist-ff-clover.log', 'r') as fp:
    for line in fp:
        if ', loss:' in line:
            mnist_ff_loss.append(float(line.split(' ')[-1]))
        elif ', acc:' in line:
            mnist_ff_accuracy.append(float(line.split(' ')[-1].split('#')[0]))


## mnist clover baseline
mnist_baseline_loss = []
mnist_baseline_accuracy = []
with open('logs/202433-mnist-clover-baseline.log', 'r') as fp:
    for line in fp:
        if 'accuracy:' in line:
            mnist_baseline_accuracy.append(float(line.split(' ')[-1]))
        else:
            mnist_baseline_loss.append(float(line) / 2)

# CIFAR10-VGG
## cifar10 clover finite field
cifar10_ff_loss = []
cifar10_ff_accuracy = []
with open('logs/202432-21315-cifar10-vgg-ff-clover.log', 'r') as fp:
    for line in fp:
        if ', loss:' in line:
            cifar10_ff_loss.append(float(line.split(' ')[-1]))
        elif ', acc:' in line:
            cifar10_ff_accuracy.append(float(line.split(' ')[-1].split('#')[0]))


## cifar10 clover baseline
cifar10_baseline_loss = []
cifar10_baseline_accuracy = []
with open('logs/202433-cifar10-vgg-clover-baseline.log', 'r') as fp:
    for line in fp:
        if 'accuracy:' in line:
            cifar10_baseline_accuracy.append(float(line.split(' ')[-1]))
        else:
            cifar10_baseline_loss.append(float(line) / 2)

# FashionMNIST
fashion_mnist_ff_loss = []
fashion_mnist_ff_accuracy = []
with open('logs/202434-2933-fashion-mnist-ff-clover.log', 'r') as fp:
    for line in fp:
        if ', loss:' in line:
            fashion_mnist_ff_loss.append(float(line.split(' ')[-1]))
        elif ', acc:' in line:
            fashion_mnist_ff_accuracy.append(float(line.split(' ')[-1].split('#')[0]))


## fashion_mnist clover baseline
fashion_mnist_baseline_loss = []
fashion_mnist_baseline_accuracy = []
with open('logs/202433-fashion-mnist-clover-baseline.log', 'r') as fp:
    for line in fp:
        if 'accuracy:' in line:
            fashion_mnist_baseline_accuracy.append(float(line.split(' ')[-1]))
        else:
            fashion_mnist_baseline_loss.append(float(line) / 2)
```

```python
# MNIST
matplotlib.rc('xtick', labelsize=23)
matplotlib.rc('ytick', labelsize=23)
plt.figure(figsize=(8, 7))
plt.plot(list(range(len(mnist_baseline_loss))), mnist_baseline_loss, linewidth=4)
plt.plot(list(range(len(mnist_ff_loss))), mnist_ff_loss, linewidth=4)
plt.ylabel('Loss', fontweight='bold', fontsize=25)
plt.xlabel('Number of iterations', fontweight='bold', fontsize=25)
plt.legend(['Baseline', 'CLOVER'], loc='upper right', prop={'size': 25, 'weight':'bold'})
plt.savefig('mnist_clover_baseline_loss.jpg', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8, 7))
plt.plot(list(range(len(mnist_baseline_accuracy))), mnist_baseline_accuracy, linewidth=4)
plt.plot(list(range(len(mnist_ff_accuracy))), mnist_ff_accuracy, linewidth=4)
plt.ylabel('Accuracy', fontweight='bold', fontsize=25)
plt.xlabel('Number of iterations', fontweight='bold', fontsize=25)
locs, labels = plt.xticks()
plt.xticks(locs[1:-1], [str(int(x * 10)) for x in locs[1:-1]])
# plt.title('Accuracy vs. Number of Iterations\n{} - Batch Size: {}'.format(title_dataset_name, batch_size))
plt.legend(['Baseline', 'CLOVER'], loc='lower right', prop={'size': 25, 'weight':'bold'})
plt.savefig('mnist_clover_baseline_acc.jpg', dpi=300)
plt.show()
```

```python
# CIFAR10
matplotlib.rc('xtick', labelsize=23)
matplotlib.rc('ytick', labelsize=23)
plt.figure(figsize=(8, 7))
plt.plot(list(range(len(cifar10_baseline_loss))), cifar10_baseline_loss, linewidth=4)
plt.plot(list(range(len(cifar10_ff_loss))), cifar10_ff_loss, linewidth=4)
plt.ylabel('Loss', fontweight='bold', fontsize=25)
plt.xlabel('Number of iterations', fontweight='bold', fontsize=25)
plt.legend(['Baseline', 'CLOVER'], loc='upper right', prop={'size': 25, 'weight':'bold'})
plt.savefig('cifar10_clover_baseline_loss.jpg', dpi=300)
plt.show()
```

```python
import numpy as np
import math
import matplotlib
import matplotlib.pylab as plt

# MNIST
## mnist clover finite field
mnist_ff_accuracy = []
with open('logs/202434-122233-mnist-ff-clover.log', 'r') as fp:
    for line in fp:
        if ', acc:' in line:
            mnist_ff_accuracy.append(float(line.split(' ')[-1].split('#')[0]))


## mnist clover baseline
mnist_baseline_accuracy = []
with open('logs/202434-mnist-clover-baseline.log', 'r') as fp:
    for line in fp:
        if 'accuracy:' in line:
            mnist_baseline_accuracy.append(float(line.split(' ')[-1]))

## mnist clover baseline
mnist_baseline_accuracy_fg = []
with open('logs/2024328-mnist-clover-baseline.log', 'r') as fp:
    for line in fp:
        if 'accuracy:' in line:
            mnist_baseline_accuracy_fg.append(float(line.split(' ')[-1]))

fig, ax = plt.subplots(figsize=(8,6), dpi=600)
for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(2)
    ax.spines[axis].set_color("black")
    ax.spines[axis].set_zorder(0)
plt.plot(mnist_ff_accuracy,color='b', linestyle='-', marker='^',markevery=2, markersize=10)
plt.plot(mnist_baseline_accuracy,color='r',linestyle='-', markevery=2, marker='X', markersize=10)
plt.plot(mnist_baseline_accuracy_fg,color='g',linestyle='-', markevery=2, marker='>', markersize=10)
plt.xlabel('# Iterations\n',fontweight='bold', fontsize=24)
plt.ylabel('Accuracy',fontweight='bold', fontsize=24)
plt.grid()
plt.rc('xtick', labelsize=23) 
plt.rc('ytick', labelsize=23) 
lgnd = plt.legend(['CLOVER', 'Conventional Neural Network'], fontsize=21)
plt.savefig("mnist_clover_baseline_acc_2.png", bbox_inches = 'tight')
plt.savefig("mnist_clover_baseline_acc_2.eps", format='eps', bbox_inches = 'tight')
plt.show()

# CIFAR10
## cifar10 clover finite field
cifar10_ff_accuracy = []
with open('logs/202434-214946-cifar10-vgg-ff-clover.log', 'r') as fp:
    for line in fp:
        if ', acc:' in line:
            cifar10_ff_accuracy.append(float(line.split(' ')[-1].split('#')[0]))


## cifar10 clover baseline
cifar10_baseline_accuracy = []
with open('logs/202434-cifar10-vgg-clover-baseline.log', 'r') as fp:
    for line in fp:
        if 'accuracy:' in line:
            cifar10_baseline_accuracy.append(float(line.split(' ')[-1]))

cifar10_baseline_accuracy_fg = []
with open('logs/2024328-cifar10-vgg-clover-baseline.log', 'r') as fp:
    for line in fp:
        if 'accuracy:' in line:
            cifar10_baseline_accuracy_fg.append(float(line.split(' ')[-1]))

fig, ax = plt.subplots(figsize=(8,6), dpi=300)
for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(2)
    ax.spines[axis].set_color("black")
    ax.spines[axis].set_zorder(0)
plt.plot(cifar10_ff_accuracy,color='b', linestyle='-', marker='^',markevery=2, markersize=10)
plt.plot(cifar10_baseline_accuracy,color='r',linestyle='-', markevery=2, marker='X', markersize=10)
plt.plot(cifar10_baseline_accuracy_fg,color='g',linestyle='-', markevery=2, marker='>', markersize=10)
plt.xlabel('# Iterations\n',fontweight='bold', fontsize=24)
plt.ylabel('Accuracy',fontweight='bold', fontsize=24)
plt.grid()
plt.rc('xtick', labelsize=23) 
plt.rc('ytick', labelsize=23) 
lgnd = plt.legend(['CLOVER', 'Conventional Neural Network (Batch GD)', 'Conventional Neural Network (Full GD)'], fontsize=21)
plt.savefig("cifar_clover_baseline_acc_2.png", bbox_inches = 'tight')
plt.savefig("cifar_clover_baseline_acc_2.eps", format='eps', bbox_inches = 'tight')
plt.show()
```
