```python
import matplotlib.pyplot as plt
```

```python
running_loss, running_acc = [], []
with open('2023517-105126-scaled-vectorized-ff-numpy.log', 'r') as fp:
    for line in fp:
        line = line[:-1]
        if line[10] == '#':
            running_loss.append(float(line.split(',')[1][10:]))
            running_acc.append(float(line.split(',')[2][5:-13]))
```

```python
plt.figure(figsize=(20, 5))
plt.plot(range(len(running_loss)), running_loss)
plt.title('loss vs. iteration - vectorized scaled finite field net')
plt.xlabel('iteration')
plt.ylabel('loss')
plt.savefig('loss_vectorized_scaled_ff.jpeg', dpi=300)
plt.show()
plt.figure(figsize=(20, 5))
plt.plot(range(len(running_acc)), running_acc)
plt.title('acc vs. iteration - vectorized scaled finite field net')
plt.xlabel('iteration')
plt.ylabel('acc')
plt.savefig('acc_vectorized_scaled_ff.jpeg', dpi=300)
plt.show()
```
